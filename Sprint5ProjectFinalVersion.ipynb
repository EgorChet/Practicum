–û–±–Ω–∞—Ä—É–∂–µ–Ω–Ω—ã–µ –∞–Ω–æ–º–∞–ª–∏–∏ —Ç—Ä–µ–±—É—é—Ç –±–æ–ª–µ–µ –¥–µ—Ç–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–∏—è –∏, –≤–æ–∑–º–æ–∂–Ω–æ, –∫–æ—Ä—Ä–µ–∫—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö.







```python
import pandas as pd

def remove_outliers(data, column, iqr_multiplier):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (iqr_multiplier * IQR)
    upper_bound = Q3 + (iqr_multiplier * IQR)
    outliers_removed = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return outliers_removed

columns_to_check = ['living_area', 'total_area', 'kitchen_area', 'last_price', 'rooms', 'ceiling_height', 'price_m2']

for column in columns_to_check:
    data = remove_outliers(data, column, 3)
```
